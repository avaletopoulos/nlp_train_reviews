# Preprocessing data

```python
def words_frequency(texts: list, terms = 10):
    words = nltk.tokenize.word_tokenize(' '.join([text for text in texts]))
    fdist = FreqDist(words)
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())})

    #Return the first terms rows ordered by columns in descending order
    d = words_df.nlargest(terms, "count") 
    plt.figure(figsize=(20,10))
    ax = sns.barplot(data=d, x= "word", y = "count")
    ax.set(ylabel = 'Count')
    ax.set(xlabel = 'Words')
    plt.show()
```

```python
stop_words = stopwords.words('english')
```

```python
df['text'] 
```

```python
# remove unwanted characters, numbers and symbols
df['text'] = df['text'].str.replace("[^a-zA-Z#]", " ")

# remove short words (length < 3)
df['text'] = df['text'].apply(lambda x: ' '.join([w for w in nltk.tokenize.word_tokenize(x) if len(w)>2]))

# remove stopwords from the text
df['text'] = df['text'].apply(lambda x: ' '.join([w for w in nltk.tokenize.word_tokenize(x) if w not in stop_words]))

# make entire text lowercase
df['text'] = df['text'].apply(lambda x: ' '.join([w.lower() for w in nltk.tokenize.word_tokenize(x)]))
```

```python
stemmer = SnowballStemmer("english")

def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

# Tokenize and lemmatize
def preprocess(text):
    result=[]
    for token in nltk.tokenize.word_tokenize(text) :
        print(token)
        result.append(lemmatize_stemming(token))
            
    return result
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

def lemmatization(texts, tags=['NOUN', 'ADJ']): # filter noun and adjective
    output = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        output.append([token.lemma_ for token in doc if token.pos_ in tags])
    return output
```

```python
tokenized_reviews = pd.Series([w for w in df['text']]).apply(lambda x: x.split())
print(tokenized_reviews[1])
```

```python
tokenized_reviews
```

```python
reviews_2 = lemmatization(tokenized_reviews)
print(reviews_2[1]) # print lemmatized review
```

```python
reviews_3 = []
for i in range(len(reviews_2)):
    reviews_3.append(' '.join(reviews_2[i]))

df['reviews'] = reviews_3

words_frequency(df['reviews'], 5)
```

```python
reviews_2
```

```python
dictionary = corpora.Dictionary(reviews_2)

doc_term_matrix = [dictionary.doc2bow(rev) for rev in reviews_2]
```

```python
doc_term_matrix
```

```python
# Creating the object for LDA model using gensim library
LDA = gensim.models.ldamodel.LdaModel

# Build LDA model
lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=3, random_state=100,
                chunksize=1000, passes=50)
```

```python
lda_model.print_topics()
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
